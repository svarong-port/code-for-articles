```python
# Import package
import polars as pl
```

## 4 Select

```python
# Select rows with []
df[:10]
```

## 5 Filter

### 5.1 Single Condition

```python
# Filter with 1 condition
df.filter(pl.col("category") == "Outdoor furniture")
```

### 5.2 Multiple Conditions

```python
# Filter with multiple conditions
df.filter(
    (pl.col("category") == "Outdoor furniture") &
    (pl.col("price") > 1000)
)
```

## 6 Sort

### 6.1 Ascending

```python
# Sort, ascending
df.sort("price")
```

### 6.2 Descending

```python
# Sort, descending
df.sort("price", descending=True)
```

### 6.3 Sort With Multiple Columns

```python
# Sort with multiple columns
df.sort(
    ["category", "price"],
    descending=[False, True]
)
```

## 7 Aggregate

### 7.1 Basic

```python
# Aggregate at dataset level
df.select(
    pl.col("price").mean().alias("Mean"),
    pl.col("price").min().alias("Min"),
    pl.col("price").max().alias("Max")
)
```

### 7.2 Group By

```python
# Aggregate with group_by
df.group_by("category").agg(
    pl.col("price").mean().alias("Mean"),
    pl.col("price").min().alias("Min"),
    pl.col("price").max().alias("Max")
)
```

## 8 Mutate

### 8.1 Add Columns

```python
# Add new columns
df.with_columns(
    discount = pl.when(pl.col("price") > 1000)
    .then(0.15)
    .otherwise(0.10),
).with_columns(
    price_discounted = pl.col("price") * (1 - pl.col("discount"))
)
```

### 8.2 Remove Columns

```python
# Remove column
df.drop(["old_price", "sellable_online"])
```

## 9 Lazy Execution

```python
# Create lazy df
df_lz = df.lazy()
```

```python
# Lazy execution
execution = df_lz.select(["name", "category", "price"])
```

```python
# Execute
execution.collect()
```

## 10 Chaining

```python
# Chaining
df_lz.filter(
    pl.col("designer") == "Francis Cayouette"
).group_by(
    "category"
).agg(
    pl.col("price").mean().round().alias("avg_price")
).with_columns(
    pl.when(pl.col("avg_price") > 1000)
    .then(pl.lit("Premium"))
    .otherwise(pl.lit("Affordable"))
    .alias("price_label")
).sort(
    "avg_price",
    descending=True
).select(
    [
        "category",
        "price_label",
        "avg_price"
    ]
).collect()
```
