## 1. Importing Libraries

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Loading Built-In Dataset

```python
# Load the dataset
tips = sns.load_dataset("tips")
```

## 3. Previewing the Dataset

```python
# Preview the datasets

# tips
tips.head()
```

## 4. Creating Basic Graphs

### 4.1 Histograms

```python
# 1. Histogram

# Create a histogram of tips
sns.histplot(data = tips,
             x = "tip")

# Show the plot
plt.show()
```

### 4.2 Box Plots

```python
# 2. Box plot

# Create a box plot of tips by time
sns.boxplot(data = tips,
            x = "time",
            y = "tip")

# Show the plot
plt.show()
```

### 4.3 Scatter Plots

```python
# 3. Scatter plot

# Create a scatter plot of tips vs total bill
sns.scatterplot(data = tips,
                x = "total_bill",
                y = "tip")

# Show the plot
plt.show()
```

### 4.4 Line Plots

```python
# 4. Line plot

# Create a line plot of tips vs party size
sns.lineplot(data = tips,
             x = "size",
             y = "tip")

# Show the plot
plt.show()
```

### 4.5 Bar Plots

```python
# 5. Bar plot

# Create a bar plot of tips vs day of week
sns.barplot(data = tips,
            x = "day",
            y = "tip")

# Show the plot
plt.show()
```

## 5. Adding a Third Variable

```python
# Scatter plot

# Create a scatter plot: tips vs total bill vs smoker types
sns.scatterplot(data = tips,
                x = "total_bill",
                y = "tip",
                hue = "smoker")

# Show the plot
plt.show()
```

## 6. Customising Graphs

### 6.1 Colours

```python
# Customise colours with `palette`

# Specify colours
colours = {"Yes": "red",
           "No": "black"}

# Create a scatter plot
sns.scatterplot(data = tips,
                x = "total_bill",
                y = "tip",
                hue = "smoker",
                palette = colours)

# Show the plot
plt.show()
```

```python
# Customise colours with `sns.set_palette()`

# Set the palette
sns.set_palette("colorblind")

# Create a scatter plot
sns.scatterplot(data = tips,
                x = "total_bill",
                y = "tip",
                hue = "smoker")

# Show the plot
plt.show()
```

### 6.2 Style

```python
# Customise style with `sns.set_style()`

# Set the style
sns.set_style("darkgrid")

# Create a scatter plot
sns.scatterplot(data = tips,
                x = "total_bill",
                y = "tip",
                hue = "smoker")

# Show the plot
plt.show()
```

### 6.3 Text

```python
# Create a scatter plot
sns.scatterplot(data = tips,
                x = "total_bill",
                y = "tip",
                hue = "smoker")

# Add a title
plt.title("Total Bill vs Tip", fontsize = 16)

# Add labels
plt.xlabel("Total Bill ($)", fontsize = 12)
plt.ylabel("Tip ($)", fontsize = 12)

# Show the plot
plt.show()
```
